```python
def sortSongsByTempo(): #this returns the 100 songs with the fastest tempo
    topTempos = musicDF.sort_values(by = 'song.tempo',ascending=False)
    topTempos = topTempos.head(100)
    return topTempos
sortSongsByTempo().plot(kind='bar',x='artist.name',y='song.tempo',figsize=(25,25)) #this shows the artist names who have made the top 100 fastest songs
```

```python
# The main question I was trying to answer: is there a correlation with the popularity of a song and its tempo (do people like faster paced or slower paced music)??
top100Speed = sortSongsByTempo().reset_index()
def findRepeatingArtists():
    count={}
    for i in range(len(top100Speed)):
        count[top100Speed['artist.name'][i]] = count[top100Speed['artist.name'][i]] + 1 if top100Speed['artist.name'][i] in count else 1
    return count
artistDF = pd.DataFrame(list(findRepeatingArtists().items()),columns = ['artist.name','freq'])
artistDF.plot(kind='bar',x='artist.name',y='freq',figsize=(25,25)) #looks like there are no repeating artists in the top 100 fastest songs
```
