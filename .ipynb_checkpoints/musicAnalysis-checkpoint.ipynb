```python
def checkForDuplicateArtists(): #this will check if we need to aggregate some sort of data
    saved=set()
    for num in musicData['artist.id']:
        if num in saved:
            return True
        saved.add(num)
    return False
```

```python
def aggregateArtistHottnessById():
    #this will return the aggregate artist hotttnesss grouped by artist id returned in a dictionary object
    hotSumById={}
    ids = musicData['artist.id']
    hottness = musicData['artist.hotttnesss']
    for i in range(len(ids)):
        if ids[i] in hotSumById:
            hotSumById[ids[i]] = hotSumById[ids[i]] + hottness[i]
        else:
            hotSumById[ids[i]] = hottness[i]
    return hotSumById
```

```python
aggregateArtistHottnessById()
```
