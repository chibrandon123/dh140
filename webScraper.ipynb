```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
response = requests.get("https://corgis-edu.github.io/corgis/csv/")
html_text = response.text
doc = BeautifulSoup(html_text, "html.parser")
datasetTopicsList = doc.find_all('h3')
```

```python
def grabDataDF(name):
    # assuming that string passed in will have the <h3> tags removed from list
    newResponse = requests.get("https://corgis-edu.github.io/corgis/csv/" + name.lower())
    tempHtmlTxt = newResponse.text
    newDoc = BeautifulSoup(tempHtmlTxt,"html.parser")
    urlString = newDoc.find('a',attrs = {'download':True}).attrs['href']
    corgisdf = pd.read_csv('https://corgis-edu.github.io/corgis/' + urlString.split('..')[2])
    return corgisdf
```

```python
billionairesDF = grabDataDF('Billionaires')
#print(billionairesDF)
billionairesDF=billionairesDF.filter(items=['name','rank']).drop_duplicates()[:20]
billionairesDF.plot(title = "Top 20 ranked Billionaires", kind = 'bar',x = "name", xlabel="name",ylabel = "rankings")
```

```python
drugsDF = grabDataDF('Drugs').filter(items = ['State','Totals.Alcohol.Use Disorder Past Year.12-17','Year'])
drugsDF = drugsDF[drugsDF.State == 'California']
drugsDF.plot(title = "Alcohol Use Disorder in California over Time", kind = "line",ylabel = "Alcohol Use Disorder", x = "Year", y = "Totals.Alcohol.Use Disorder Past Year.12-17")
```

```python
healthDF = grabDataDF('Health')
healthDF = healthDF.filter(items = ['disease','loc','year'])
healthDF = healthDF[healthDF['disease'] == 'MEASLES']
healthDF.plot(kind = 'hist',xlabel="Year", title = "Amount of Measles Outbreaks per Year")
```
